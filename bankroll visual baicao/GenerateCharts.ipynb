```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu (Thay tên file CSV tương ứng)
# Giả sử file có cột: 'Round', 'AI_1', 'AI_2', ..., 'AI_17'
df = pd.read_csv('20260513_232354_bankroll_history_deuca3.csv')

# 2. Setup phong cách biểu đồ
plt.figure(figsize=(14, 7))
sns.set_style("darkgrid")

# 3. Lọc ra những con giàu nhất và nghèo nhất (dựa vào số dư ở ván cuối cùng)
final_balances = df.iloc[-1, 1:] # Bỏ cột Round
top_ais = final_balances.nlargest(3).index.tolist()
bottom_ais = final_balances.nsmallest(3).index.tolist()

# 4. Vẽ biểu đồ
for col in df.columns[1:]:
    if col in top_ais:
        # Nhóm Shark chiến thắng: Xanh lá, nét đậm
        plt.plot(df['Round'], df[col], label=f'Top Winners ({col})', color='#2ecc71', linewidth=2.5, alpha=0.9)
    elif col in bottom_ais:
        # Nhóm Phá sản: Đỏ, nét đậm
        plt.plot(df['Round'], df[col], label=f'Bottom Losers ({col})', color='#e74c3c', linewidth=2.5, alpha=0.9)
    else:
        # Đám đông còn lại: Xám, nét mỏng, mờ
        plt.plot(df['Round'], df[col], color='grey', linewidth=1, alpha=0.2)

# Xử lý Legend (chống trùng lặp)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='upper left', fontsize=12)

# 5. Trang trí tiêu đề
plt.title('Bankroll History: Sự phân hóa tài sản sau 50,000 ván (Hồ Hỗn Tạp)', fontsize=16, fontweight='bold')
plt.xlabel('Số ván đấu (Rounds)', fontsize=12)
plt.ylabel('Credit Balance (Khởi điểm 10,000)', fontsize=12)
plt.axhline(y=10000, color='black', linestyle='--', linewidth=1.5, label='Vốn ban đầu')

plt.tight_layout()
plt.savefig('bankroll_visualization.png', dpi=300)
print("Đã lưu biểu đồ thành công!")
```
